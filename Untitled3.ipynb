```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv("casasboston.csv")
```

```python
print(datos)
```

```python
df=datos.rename(columns={
    "TOWN":"Ciudad",
    "CRIM":"Indice",
    "AGE":"Edad",
    "TAX":"TAX",
    "MEDV":"VALOR_MEDIANO",
    "INDUS":"Zona_Industrial",
    "RM":"N_Habitantes",
})
```

```python
df.N_Habitantes.plot.hist()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,4),facecolor='blue',
                       layout='constrained')
fig.suptitle('BARRAS')
ax.set_title('Media De Habitantes', loc='center',fontstyle='oblique', fontsize='large')
df.N_Habitantes.plot.hist()
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO",y= "Indice",alpha=0.7)
plt.show()
```

```python
valor_por_ciudad=df.groupby("Ciudad")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh()
plt.show()
```
